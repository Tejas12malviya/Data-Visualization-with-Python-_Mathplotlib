```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
# LINE PLOT
```

```python
x=np.arange(-10,11)
x
```

```python
y=x**2
y
```

```python
plt.plot(x,y,color='r',linestyle=':',linewidth=3)
plt.title("Line Plot")
plt.xlabel("x-label")
plt.ylabel("y-label")
plt.show()
```

```python
x1=np.arange(-7,8)
y1=x1**2
y2=2*x1**2
plt.plot(x1,y1,color='g',linestyle=':',linewidth=2.5)
plt.plot(x1,y2,color='r',linestyle='-.',linewidth=2.5)
plt.title("Line Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)
plt.show()
```

```python
x=np.arange(-7,8)
y1=x**2
y2=2*x**2
plt.subplot(1,2,1)
plt.plot(x1,y1,color='g',linestyle=':',linewidth=2.5)
plt.title("Line Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)

plt.subplot(1,2,2)
plt.plot(x1,y2,color='r',linestyle='-.',linewidth=2.5)
plt.title("Line Plot")
plt.xlabel("x-axis")
plt.grid(True)
plt.show()
```

```python
# BAR PLOT
```

```python
student= {"Bob":87,"Matt":56,"Sam":27,'Anna':50}
```

```python
names= list(student.keys())
values= list(student.values())
plt.bar(names,values)
plt.title('Distribution Of Students Marks')
plt.xlabel('Names of Students')
plt.ylabel('Marks')
plt.show()
```

```python
# Horizontal Bar Graph
plt.barh(names,values,color="g")
plt.title('Bar Plot')
plt.xlabel('Names Of Students')
plt.ylabel('Marks')
plt.grid(True)
plt.show()
```

```python
# Scatter Plot
```

```python
x=[10,20,30,40,50,60,70,80,90]
a=[8,2,4,1,9,6,7,3,5]

plt.scatter(x,a,marker="*",c="r",s=60)
plt.grid(True)
plt.show()
```

```python
x=[10,20,30,40,50,60,70,80,90]
a=[8,2,4,1,9,6,7,3,5]
b=[4,3,7,2,5,9,8,1,6]

plt.scatter(x,a,marker="*",c="r",s=60)
plt.scatter(x,b,marker=".",c="g",s=80)

plt.show()
```

```python
x=[10,20,30,40,50,60,70,80,90]
a=[8,2,4,1,9,6,7,3,5]
b=[4,3,7,2,5,9,8,1,6]

plt.subplot(1,2,1)
plt.scatter(x,a,marker="*",c="r",s=60)

plt.subplot(1,2,2)
plt.scatter(x,b,marker=".",c="g",s=80)

plt.show()
```

```python
# Box Plot
```

```python
one=[1,2,3,4,5,6,7,8,9]
two=[1,2,3,4,5,4,3,2,1]
three=[6,7,8,9,8,7,6,5,4]

data=list([one,two,three])
plt.boxplot(data)
plt.show()
```

```python
one=[1,2,3,4,5,6,7,8,9]
two=[1,2,3,4,5,4,3,2,1]
three=[6,7,8,9,8,7,6,5,4]

data=list([one,two,three])
plt.violinplot(data)
plt.show()
```

```python
# Pie Chart
```

```python
fruit=['Apple','Orange','Mango','Guava','Pineapple']
quantity = [67,34,100,45,38]

plt.pie(quantity,labels=fruit,autopct='%0.1f%%',colors=['red','orange','blue','green','yellow'])
plt.show()
```

```python
# Doughnut Chart
```

```python
fruit=['Apple','Orange','Mango','Guava','Pineapple']
quantity = [67,34,100,45,38]

plt.pie(quantity,labels=fruit,radius=2,autopct='%0.1f%%',colors=['red','orange','blue','green','yellow'])
plt.pie([1],colors=['w'],radius=1)
plt.show()
```
